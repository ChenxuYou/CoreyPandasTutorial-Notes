```python
python_percentage = country_grp['LanguageWorkedWith'].apply(
    lambda x: x.str.contains('Python').sum()/x.size*100 if x.size > 0 else 0)
```

```python
countries_use_python = country_grp['LanguageWorkedWith'].apply(
    lambda x: x.str.contains('Python').sum())
countries_use_python
```

```python
countries_respondents = df['Country'].value_counts()
countries_respondents
```

```python
python_df = pd.concat(
    [countries_respondents, countries_use_python], axis='columns', sort=False)
python_df
```

```python
python_df.rename(columns={'LanguageWorkedWith': 'NumUsePython',
                 'count': 'NumRespondents'}, inplace=True)
```

```python
python_df['Percentage'] = (
    python_df['NumUsePython'] / python_df['NumRespondents']) * 100
python_df
```

```python
python_df.sort_values(by=['Percentage'], ascending=False).head(15)
```
